## Visulation

```python
total = len(data)
ax = sbn.countplot(x=data["Sex"])
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax.annotate(percentage, (x, y),ha='center')
plt.show()
```

```python
total = len(data)
ax = sbn.barplot(x = "Sex", y = "Job", data = data)
```

```python
plt.figure(figsize=(15,8))
total = len(data)
ax = sbn.countplot(x=data["Purpose"])
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax.annotate(percentage, (x, y),ha='center')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sbn.barplot(x = "Purpose",y = "Credit amount",data=data)
```

```python
def scatters(data , h=None):
    fig, (ax1,ax2,ax3) = plt.subplots(3,1, figsize=(8,8))
    sbn.scatterplot(x="Credit amount", y="Duration",data=data,hue=h,ax=ax1)
    sbn.scatterplot(x="Age", y="Credit amount",data=data,hue=h,ax=ax2)
    sbn.scatterplot(x="Age", y="Duration",data=data,hue=h,ax=ax3)
    plt.tight_layout()
    
scatters(data,"Sex")
    
```

```python
def boxes(x,y,h,r=45):
    fig, ax = plt.subplots(figsize=(10,6))
    box = sbn.boxplot(x=x,y=y, hue=h, data=data)
    box.set_xticklabels(box.get_xticklabels(), rotation=r)
    fig.subplots_adjust(bottom=0.2)
    plt.tight_layout()
```

```python
boxes("Purpose","Credit amount","Sex")
```

```python
boxes("Purpose","Duration","Sex")
```

```python
boxes("Housing","Credit amount","Sex",r=0)
```

```python
boxes("Job","Credit amount","Sex",r=0)
```
