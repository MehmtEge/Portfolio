```python
import nbconvert
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
import warnings
warnings.filterwarnings("ignore")
```

## Excel dosyamızı okutuyoruz

```python
dataFrame = pd.read_csv("merc.xlsx")
```

## Verimiz hakkında bilgi edinmek için ilk 5 satır

```python
dataFrame.head()
```

## Verimizin ortalamasını,standart sapmasını,in ve max gibi değerleri

```python
dataFrame.describe()
```

## Kullanıcıların tercih ettiği vites tipine bakıyoruz

```python
sbn.countplot(dataFrame["transmission"])
plt.show()
```

## Her modelden kaç adet alındığın

```python
print(dataFrame["model"].value_counts())
print("---------------------------------------------")
print(dataFrame["model"].value_counts() / len(dataFrame))
sbn.countplot( y = dataFrame["model"])
plt.show()
```

## Modele göre mile başına yakıt tüketim

```python
sbn.barplot(dataFrame["mpg"],dataFrame["model"])
plt.show()
```

## Yıla göre yakıt tüketimini

```python
plt.figure(figsize = (15,8))
ax = sbn.barplot(x = dataFrame["year"],y = dataFrame["mpg"], capsize=.2)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
plt.show()
```

## Yıla göre ödenen ortalama vergi

```python
plt.figure(figsize=(15,8))
ax = sbn.barplot(x = dataFrame["year"], y = dataFrame["tax"], capsize = .2)
ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
plt.show()
```

## Yıla göre ortalama fiyat

```python
plt.figure(figsize = (15,8))
ax = sbn.barplot(dataFrame["year"],dataFrame["price"])
ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
plt.show()
```

## Yapılan kilometerye göre araç fiyatı

```python
plt.figure(figsize = (15,8))
sbn.scatterplot(dataFrame["mileage"],dataFrame["price"])
```

## Yapılan kilometerye göre araç fiyatı - yakıt türüne göre gösterim

```python
plt.figure(figsize = (15,8))
sbn.scatterplot(dataFrame["mileage"],dataFrame["price"],hue=dataFrame["fuelType"])
```

## Vites türüne göre fiyat aralığı

```python
plt.figure(figsize = (15,8))
sbn.barplot(dataFrame["transmission"],dataFrame["price"])
```

## Araç fiyatına göre vergi tutarı

```python
plt.figure(figsize = (15,8))
ax = sbn.scatterplot(x = dataFrame["price"],y = dataFrame["tax"])
plt.show()
```

## Yıla göre vargi tutarı

```python
plt.figure(figsize = (15,8))
ax = sbn.scatterplot(x = dataFrame["year"],y = dataFrame["tax"])
plt.show()
```

## Yakıt türüne göre mile başına yakıt tüketimi

```python
plt.figure(figsize = (15,8))
sbn.violinplot(dataFrame["fuelType"],dataFrame["mpg"],hue=dataFrame["fuelType"])
plt.show()
```

## Tüm değerleri birbiri ile olan ilşkisi
